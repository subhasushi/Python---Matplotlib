```python
#import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import csv
import os
import numpy as np
```

```python
#read csv files
csv_path = os.path.join("raw_data","city_data.csv")
city_data = pd.read_csv(csv_path)
city_data.head()
```

```python
#read csv files
csv_path1 = os.path.join("raw_data","ride_data.csv")
ride_data = pd.read_csv(csv_path1)
ride_data.head()
```

```python
#merging 2 data sets
combined_data = pd.merge(city_data,ride_data, on = "city")
combined_data
```

```python
#grouping data by type (Urban)
combined_data_urban = combined_data.loc[combined_data["type"] == "Urban", :]
combined_data_urban.head()
```

```python
#Calculating total rides per city for Urban type
urban_rides_percity = combined_data_urban["city"].value_counts()
urban_rides_percity.head()
```

```python
#Calculating total driver_count per city for Urban type
driver_count_percity = combined_data_urban[["city","driver_count"]].drop_duplicates().set_index('city')
driver_count_percity.head()
```

```python
#calculate average fare per city - Type (Urban)
avg_fare_percity = combined_data_urban[["city","fare"]].drop_duplicates()
avg_fare_percity = round(avg_fare_percity.groupby(["city"]).mean(),2)
avg_fare_percity.head()
```

```python
#Summary of Urban data
urban_summary = pd.DataFrame({"Total No of Rides Per City": urban_rides_percity,
                              "Avg fare per city" : avg_fare_percity['fare'],"Driver count per city":driver_count_percity['driver_count']})
urban_summary.head()
```

```python
#grouping data by type (Suburban)
combined_data_suburban = combined_data.loc[combined_data["type"] == "Suburban", :]
combined_data_suburban.head()
```

```python
#calculating total rides per city for Suburban type
suburban_rides_percity = combined_data_suburban["city"].value_counts()
suburban_rides_percity.head()
```

```python
#calculating driver_count per city suburban type
driver_count_percity_suburban_raw = combined_data_suburban[["city","driver_count"]].drop_duplicates().set_index('city')
driver_count_percity_suburban = driver_count_percity_suburban_raw.groupby(["city"]).sum()

```

```python
#calculate average fare per city - Type (Suburban)
avg_fare_percity_suburban = combined_data_suburban[["city","fare"]].drop_duplicates()
avg_fare_percity_suburban = round(avg_fare_percity_suburban.groupby(["city"]).mean(),2)
avg_fare_percity_suburban.head()
```

```python
#summary of suburban data
suburban_summary = pd.DataFrame({"Total No of Rides Per City": suburban_rides_percity,
                              "Avg fare per city" : avg_fare_percity_suburban['fare'],
                            "Driver count per city":driver_count_percity_suburban['driver_count']})
suburban_summary.head()
```

```python
#grouped data for rural type
combined_data_rural = combined_data.loc[combined_data["type"] == "Rural", :]
combined_data_rural.head()
```

```python
#total rides per city for rural type
rural_rides_percity = combined_data_rural["city"].value_counts()
rural_rides_percity.head()
rural_rides_percity.head()
```

```python
#total driver_count per city for rural type
driver_count_percity_rural_raw = combined_data_rural[["city","driver_count"]].drop_duplicates().set_index('city')
driver_count_percity_rural = driver_count_percity_rural_raw.groupby(["city"]).sum()

driver_count_percity_rural.head()
```

```python
#calculate average fare per city - Type (Rural)
avg_fare_percity_rural = combined_data_rural[["city","fare"]].drop_duplicates()
avg_fare_percity_rural = round(avg_fare_percity_rural.groupby(["city"]).mean(),2)
avg_fare_percity_rural.head()
```

```python
#summary of rural data
rural_summary = pd.DataFrame({"Total No of Rides Per City": rural_rides_percity,
                              "Avg fare per city" : avg_fare_percity_rural['fare'],
                            "Driver count per city":driver_count_percity_rural['driver_count']})
rural_summary.head()
```

```python
#plotting bubble chart for pyber 

plt.scatter(x = urban_summary["Total No of Rides Per City"], y = urban_summary["Avg fare per city"], s = urban_summary["Driver count per city"]*5,label='Urban', color = "Lightcoral", edgecolor = "black", alpha = 0.75)
plt.scatter(x = suburban_summary["Total No of Rides Per City"], y = suburban_summary["Avg fare per city"], s = suburban_summary["Driver count per city"]*5 ,label='Suburban', color = "Lightskyblue", edgecolor = "black", alpha = 0.75)
plt.scatter(x = rural_summary["Total No of Rides Per City"], y =rural_summary["Avg fare per city"], s = rural_summary["Driver count per city"]*5 , label="Rural",color = "gold", edgecolor = "black", alpha = 0.75)

#labelling the axes and giving title name
plt.xlabel("Total No of Rides per City")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016) ")

#adding annotation
plt.text(45, 35, "Note:")
plt.text(45,33,"Circle size corresponds to driver_count per city")

#setting x and y limits
plt.xlim(0,40)
plt.ylim(15,45)

#fixing same size for the points in the legend
lgnd = plt.legend(loc="best", scatterpoints=1, fontsize=10)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

#plotting the chart
plt.show()
```

```python
#calculate total fare of all cities
grouped_data = combined_data.groupby(["city"])
totalfare_city = grouped_data["fare"].sum()
total_fare = round(totalfare_city.sum(),2)
total_fare
```

```python
#Calculate total fares by each type
total_fare_type = combined_data.groupby(["type"])["fare"].sum()
total_fare_type_df = pd.DataFrame(total_fare_type)
total_fare_type_df
```

```python
#percentage total fare (Rural)
per_totalfare_rural = round((total_fare_type_df.loc["Rural"]/total_fare)*100,2)
per_totalfare_rural
```

```python
#percentage total fare (Suburban)
per_totalfare_suburban = round((total_fare_type_df.loc["Suburban"]/total_fare)*100,2)
per_totalfare_suburban
```

```python
#percentage total fare (Urban)
per_totalfare_urban = round((total_fare_type_df.loc["Urban"]/total_fare)*100,2)
per_totalfare_urban
```

```python
#Total rides per city
total_rides_percity = combined_data.groupby(["type","city"])['ride_id'].count()
total_rides = total_rides_percity.sum()
total_rides
```

```python
#Total rides per by type
total_rides_type = combined_data.groupby(["type"])["ride_id"].count()
total_rides_type_df = pd.DataFrame(total_rides_type)
total_rides_type_df 
```

```python
#Percentage Total rides (Rural)
per_rides_rural = round((total_rides_type_df .loc["Rural"]/total_rides)*100,2)
per_rides_rural
```

```python
#Percentage Total rides (Suburban)
per_rides_suburban = round((total_rides_type_df .loc["Suburban"]/total_rides)*100,2)
per_rides_suburban
```

```python
#Percentage Total rides (Urban)
per_rides_urban = round((total_rides_type_df .loc["Urban"]/total_rides)*100,2)
per_rides_urban
```

```python
#Total driver_count for all cities
total_driver_count = city_data["driver_count"].sum()
total_driver_count
```

```python
#driver count by type
driver_count_type = city_data.groupby(["type"])["driver_count"].sum()
driver_count_type_df = pd.DataFrame(driver_count_type)
driver_count_type_df
```

```python
#Percentage driver count (Rural)
per_drivercount_rural = round((driver_count_type_df.loc["Rural"]/total_driver_count)*100,2)
per_drivercount_rural 
```

```python
#Percentage driver count (Suburban)
per_drivercount_suburban = round((driver_count_type_df.loc["Suburban"]/total_driver_count)*100,2)
per_drivercount_suburban
```

```python
#Percentage driver count (urban)
per_drivercount_urban = round((driver_count_type_df.loc["Urban"]/total_driver_count)*100,2)
per_drivercount_urban 
```

```python
# Pie Chart for % of Total Fares by City Type
# Labels for the sections of our pie chart

labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [per_totalfare_rural, per_totalfare_suburban, per_totalfare_urban]

# The colors of each section of the pie chart
colors = ["Gold", "Lightskyblue", "Lightcoral"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title('% Total Drivers by City Type')
plt.show()
```

```python
# Pie Chart for % of Total Rides by City Type
# Labels for the sections of our pie chart

labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [per_rides_rural, per_rides_suburban, per_rides_urban]

# The colors of each section of the pie chart
colors = ["Gold", "Lightskyblue", "Lightcoral"]

plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title('% Total Rides by City Type')
plt.show()
```

```python
# Pie Chart for % of Total Drivers by City Type
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [per_drivercount_rural, per_drivercount_suburban, per_drivercount_urban]

# The colors of each section of the pie chart
colors = ["Gold", "Lightskyblue", "Lightcoral"]

plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% Total Drivers by City Type")
plt.show()
```
